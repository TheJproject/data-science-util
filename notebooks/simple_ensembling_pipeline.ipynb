```python
import pandas as pd
import tqdm
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier,StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from omegaconf import OmegaConf
```

```python
from omegaconf import OmegaConf
# loading
config = OmegaConf.load(r'C:\Users\Jonah\OneDrive\Documents\Thesis\data-science-util\config.yaml')
HP = config['hyperparameters']
```

```python
# load the prediction dataframes produced by model1 and model2
df1 = pd.read_csv(r"C:\Users\Jonah\OneDrive\Documents\Thesis\data-science-util\data\raw\train_dev.csv", index_col=False)

df1 = df1.set_index('Id')
df2 = pd.read_csv(r"C:\Users\Jonah\OneDrive\Documents\Thesis\data-science-util\data\raw\train_dev.csv", index_col=False)
df2 = df2.set_index('Id')
```

```python
df2
```

```python
X = pd.concat([df1, df2], axis=1).iloc[:, :-1]
y = df1.iloc[:,-1]
```

```python
X
```

```python
# make sure the prediction dataframes have the same number of rows
assert df1.shape[0] == df2.shape[0], "The prediction dataframes have different number of rows"

# merge the prediction dataframes into a single dataframe

# load the true labels for the data

# split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=config['random_sate'])
```

```python
# train two base models, a Random Forest classifier and a Logistic Regression
base_RandomForestClassifier = RandomForestClassifier(n_estimators=HP['random_forest']['n_estimators'], random_state=config['random_sate'])
base_LogisticRegression = LogisticRegression(random_state=config['random_sate'])
base_AdaBoostClassifier = AdaBoostClassifier(n_estimators=HP['ada_boost']['n_estimators'],random_state=config['random_sate'])
```

```python
base_RandomForestClassifier.fit(X_train, y_train)
base_LogisticRegression.fit(X_train, y_train)
base_AdaBoostClassifier.fit(X_train, y_train)
```

```python
# make predictions with the base models
base_predictions_train = pd.DataFrame({
    'RandomForest': base_RandomForestClassifier.predict(X_train),
    'LogisticRegression': base_LogisticRegression.predict(X_train),
    'AdaBoostClassifier' : base_AdaBoostClassifier.predict(X_train)
})
```

```python
# train a second-level model using the predictions of the base models
stacker = RandomForestClassifier(n_estimators=HP['random_forest']['n_estimators'], oob_score=True, random_state=config['random_sate'])
stacker.fit(base_predictions_train, y_train)

# make predictions with the stacked model using the predictions of the base models on the test set
base_predictions_test = pd.DataFrame({
    'RandomForest': base_RandomForestClassifier.predict(X_test),
    'LogisticRegression': base_LogisticRegression.predict(X_test),
    'AdaBoostClassifier': base_AdaBoostClassifier.predict(X_test)
})
stacked_prediction = stacker.predict(base_predictions_test)
```

```python
# save the stacked prediction to a new csv file
pd.DataFrame({'Id': pd.Series(y_test.index,name='Id'), 'stacked_prediction': stacked_prediction}).to_csv("stacked_prediction.csv", index=False)
```

```python
from sklearn.metrics import cohen_kappa_score
cohen_kappa_score(y_test,stacked_prediction)
```
